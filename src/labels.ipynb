```python
# param to run the encoding pipeline on all subjects or on 1 subject (faster)
batch_mode = True
```

```python
import os
import pandas as pd
import numpy as np

subj_ids = np.load("gcn_input/subject_ids.npy")
df = pd.read_csv("abide_timeseries/phenotypic.csv")

label_dict = dict(zip(df["SUB_ID"].astype(str), df["DX_GROUP"]))

timeseries_dir = "./abide_timeseries/aal_csv/"
labels = []

for index, subj_id in enumerate(subj_ids):
    
    label = label_dict[subj_id] - 1
    labels.append(label)
        
    if not batch_mode:
        if index == 20:
            break

labels = np.array(labels)
print(labels)
print(labels.min())
print(labels.max())
np.save("gcn_input/labels.npy", labels)
```
